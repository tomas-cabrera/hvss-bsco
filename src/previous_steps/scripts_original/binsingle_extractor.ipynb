```python
%pylab inline
import cmc_parser as cp
import os
```

```python
def type_identifier(k1,k2,k3):
    """
    k1,k2 are binary components
    k3 is single
    
    returns 1 [bh,bh]+star
            2 [bh,star]+bh
            4 [bh,star]+star
            3 [star,star]+bh
            0 otherwise
    """
    if (k1 >= 10) & (k2>=10) and (k3 != 14):
        return 1
    elif (((k1 >= 10) & (k2 != 14)) | ((k1 != 14) & (k2>=10))) and (k3 >= 10):
        return 2
    elif (((k1 >= 10) & (k2 != 14)) | ((k1 != 14) & (k2 >= 10))) and (k3 != 14):
        return 3
    elif (k1 != 14) & (k2 != 14) and (k3 >= 10):
        return 4
    else:
        return 0
```

```python
files = [f+'/' for f in os.listdir('.') if 'tar.gz' in f]

for prefix in files:
    
    print(prefix)
    
#     if os.path.isfile(prefix+"full_initial.encounters.dat"):
#         continue

    binintfile=prefix+'initial.binint.log'
    unitsfile=prefix+'initial.conv.sh'

    with open(binintfile, 'r') as f:
            num_int = sum([1 for line in f if '****' in line])
            f.seek(0) #need to count the encounters then reset the file handle
            binints = [cp.binint(f) for __ in range(num_int)]

    units = cp.conversion_file(unitsfile)

    binsingles = []

    for binint in binints:
        if binint.type == 'binary-single':
            time = binint.time * units.time_myr
            
            b = binint.b
            v = binint.v
            vesc = binint.vesc

            a = binint.in_binaries[0].a
            e = binint.in_binaries[0].e
            m1 = binint.in_binaries[0].m1
            m2 = binint.in_binaries[0].m2
            k1 = binint.in_binaries[0].k1
            k2 = binint.in_binaries[0].k2
            r1 = binint.in_binaries[0].r1
            r2 = binint.in_binaries[0].r2

            k3 = binint.in_singles[0].ktype
            m3 = binint.in_singles[0].m
            r3 = binint.in_singles[0].r

            type_encounter = type_identifier(k1,k2,k3)

            if type_encounter > 0:
                binsingles.append([time,b,v,a,e,vesc,m1,m2,m3,r1,r2,r3,k1,k2,k3,type_encounter])

    binsingles = array(binsingles)            
    header='#1:time[Myr] #2:b/a #3:v/v_crit #4:a[AU] #5:e #6:vesc[km/s] #7:m1[msun] #8:m2 #9:ms \
    #10:r1[Rsun] #11:r2 #12:rs #13:ktype1 #14:ktype2 #15:ktypes #16:type[1:bhbh+s,2:bhs+bh,3:bhs+s,4:ss+bh]'

    savetxt(prefix+"full_initial.encounters.dat",binsingles,fmt="%.6g",header=header)
```
